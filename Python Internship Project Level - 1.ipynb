### Task 1 : Top Cuisines

### Determine the top three most common cuisines in the dataset.

```python
top_cuisines = dt['Cuisines'].value_counts()
```

```python
# top 3 most common cuisines
top_3_cuisines = top_cuisines.head(3)

print(f"Top three most common cuisines: {top_3_cuisines}")
```

### Calculate the percentage of restaurants that serve each of the top cuisines

```python
# the total length of the number of restaurants
number_of_restaurants = len(dt)
```

```python
# get the percentage of restaurants' each cuisines
percentage_of_restaurants = (top_3_cuisines / number_of_restaurants) * 100

print(f"The Percentage of restaurants that served each of the top cuisines: {percentage_of_restaurants}")
```

### Task 2 : City Analysis

### Identify the city with the highest number of restaurants in the dataset

```python
# get the count of the number of the restaurants in each city
city_restaurant_counts = dt.groupby('City')['Restaurant ID'].count()
city_restaurant_counts
```

```python
# Get the highest number of restaurants
max_city = city_restaurant_counts.idxmax()
print(f"The city with the highest number of restaurants is: {max_city}")
```

### Calculate the average rating for restaurants in each city.


```python
# Get the average rating for a restaurants in each city
city_avg_rating = dt.groupby('City')['Aggregate rating'].mean()
print(f"The average ratings for restaurants in each city is: {city_avg_rating}")
```

### Determine the city with the highest average rating

```python
# Identify the city with the highest average rating
max_avg_rating_city = city_avg_rating.idxmax()
print(f"The city with the highest average rating is: {max_avg_rating_city}")
```

```python
# Identifyt the highest average rating
highest_average_rating = city_avg_rating.max()
print(f"The city with the highest average rating is: {highest_average_rating:.2f}")
```

### Task 3: Price Range Distribution

### Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.


```python
price_counts = dt['Price range'].value_counts()
```

```python
# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(price_counts.index, price_counts.values, color='Green')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.xticks(price_counts.index)
plt.grid(axis='y', linestyle = '--', alpha=0.7)
plt.show()
```

### Calculate the percentage of restaurants in each price range category.


```python
# count the number of restaurants
price_counts = dt['Price range'].value_counts()
```

```python
# Calculate the percentage of each price range
price_percentage = (price_counts / price_counts.sum()) * 100

print(f"The percentage of each price range: {price_percentage}")
```
